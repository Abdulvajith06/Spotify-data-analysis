Finding the outlayers


```python
#inter quantiles range

q1=data["energy"].quantile(0.25)
q2=data["energy"].quantile(0.75)
iqr=q2-q1
```

```python
iqr
```

```python
outlayers=data[(data["energy"]<q1-1.5*iqr) | (data["energy"]>q2+1.5*iqr)]
```

```python
outlayers
```

```python
print(f"No of outlayers is :{len(outlayers)}")
```

```python
print(outlayers["song_title"])
```
